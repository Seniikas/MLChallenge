```
#Separo las columnas en tipos de variables
numericos = ['age', 'bmi',  'children']
categoricos = ['sex', 'smoker', 'region'] #si bien sex y smoker son binarias, vamos a tratarlo como variables categoricas para las luego hacer el OHE.
```

a partir de ac se van a ver varias iteraciones de pipelines con distintos modelos predictivos. Se utilizar gridsearch para buscar los mejores hiperparametros de cada modelo. 

```
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
#sencillo preprocesamiento para que las escalas de las variables numericas no afecten al SVM y el OHE para las categoricas.
transformer = make_column_transformer(
    (MinMaxScaler(), numericos),
    (OneHotEncoder(), categoricos)
)
```

```
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.svm import SVR

estimadores = [('column_transformer', transformer) ,('cluster', KMeans(max_iter=100)) , ('regg', RandomForestRegressor(random_state= 42, n_estimators=300))]
pipe = Pipeline(estimadores)
pipe
```

```
from sklearn.model_selection import GridSearchCV

param_grid = {     'cluster__n_clusters':[3,4,5],
                   'regg__max_depth' :[ 20,25,None]
            }

modelo_gscv = GridSearchCV( pipe, param_grid= param_grid, cv = 3, scoring='neg_root_mean_squared_error' , verbose= 10, n_jobs= -1)
modelo_gscv.fit(dfCustomers.drop('charges',axis = 1), dfCustomers.charges)
print(modelo_gscv.best_score_)
print(modelo_gscv.best_params_)
pd.DataFrame(modelo_gscv.cv_results_)
```

```
estimadores = [('column_transformer', transformer) , ('regg', RandomForestRegressor(random_state= 42))]
pipe = Pipeline(estimadores)
pipe
param_grid = {   
                    'regg__n_estimators': [100, 200, 300, 500],
                   'regg__max_depth' :[ 20,25, 50, 200, None]
            }

modelo_gscv = GridSearchCV( pipe, param_grid= param_grid, cv = 3, scoring='neg_root_mean_squared_error' , verbose= 10, n_jobs= -1)
modelo_gscv.fit(dfCustomers.drop('charges',axis = 1), dfCustomers.charges)
print(modelo_gscv.best_score_)
print(modelo_gscv.best_params_)
pd.DataFrame(modelo_gscv.cv_results_)
```

```
estimadores = [('column_transformer', transformer) , ('regg', RandomForestRegressor(random_state= 42))]
pipe = Pipeline(estimadores)
pipe
param_grid = {   
                    'regg__n_estimators': [100, 200, 300, 400, 500],
                   'regg__max_depth' :[ 3, 5, 10, 15,25,  None]
            }

modelo_gscv = GridSearchCV( pipe, param_grid= param_grid, cv = 3, scoring='neg_root_mean_squared_error' , verbose= 10, n_jobs= -1)
modelo_gscv.fit(dfCustomers.drop('charges',axis = 1), dfCustomers.charges)
print(modelo_gscv.best_score_)
print(modelo_gscv.best_params_)
pd.DataFrame(modelo_gscv.cv_results_)
```

```
estimadores = [('column_transformer', transformer) , ('regg', SVR())]
pipe = Pipeline(estimadores)
pipe
param_grid = {   
                    'regg__kernel': ['poly', 'rbf', 'sigmoid'],
                    'regg__gamma' :[ 'scale','auto'],
                    'regg__C': [0.1, 1, 10],
                    'regg__epsilon': [0.1, 0.2, 0.5]      
            }

modelo_gscv = GridSearchCV( pipe, param_grid= param_grid, cv = 3, scoring='neg_root_mean_squared_error' , verbose= 10, n_jobs= -1)
modelo_gscv.fit(dfCustomers.drop('charges',axis = 1), dfCustomers.charges)
print(modelo_gscv.best_score_)
print(modelo_gscv.best_params_)
pd.DataFrame(modelo_gscv.cv_results_)
```

```
from sklearn.ensemble import GradientBoostingRegressor

estimadores = [('column_transformer', transformer) , ('gbr', GradientBoostingRegressor())]
pipe = Pipeline(estimadores)
pipe
param_grid = {   
                'gbr__n_estimators': [100, 200, 300],
                'gbr__learning_rate': [0.05, 0.1, 0.2],
                'gbr__max_depth': [3, 4, 5]
            }

modelo_gscv = GridSearchCV( pipe, param_grid= param_grid, cv = 3, scoring='neg_root_mean_squared_error' , verbose= 10, n_jobs= -1)
modelo_gscv.fit(dfCustomers.drop('charges',axis = 1), dfCustomers.charges)
print(modelo_gscv.best_score_)
print(modelo_gscv.best_params_)
pd.DataFrame(modelo_gscv.cv_results_)
```
